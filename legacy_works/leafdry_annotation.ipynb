```python
##bbox 좌표 정규화 함수 정의
  ##  기존 좌표값/원본 이미지 데이터 size 비율로 0~1 사이의 값으로 정규화
  ##  모델 학습 과정에서 이미지 크기를 변경하더라도 각 bbox가 변하지 않도록 하기 위함
import os
from PIL import Image

def convert_bbox_to_yolo(bbox, img_width, img_height):

    if isinstance(bbox, str):
        bbox = eval(bbox)

    x_min, y_min, width, height = bbox
    x_center = x_min + width / 2
    y_center = y_min + height / 2
    x_center_norm = x_center / img_width
    y_center_norm = y_center / img_height
    width_norm = width / img_width
    height_norm = height / img_height
    return [x_center_norm, y_center_norm, width_norm, height_norm]

def get_image_dim(image_path):
  with Image.open(image_path) as img:
    return img.width, img.height

```

```python
## 어노테이션 생성을 위한 데이터프레임 준비
  #1) convert_bbox_to_yolo 함수를 실행하기 위해 라벨링 데이터프레임에 각 이미지 파일의 width, height 값을 포함하는 컬럼 생성
  #2) 어노테이션 생성을 위해 정규화한 bbox 값을 저장하는 yolo_bboxes 데이터프레임 생성

image_directory = '/content/leafdry/train/train_img/'
df_train_class['image_path'] = image_directory + df_train_class['img']
df_train_class['image_dimensions'] = df_train_class['image_path'].apply(get_image_dim)
df_train_class[['image_width', 'image_height']] = df_train_class['image_dimensions'].apply(pd.Series)

yolo_bboxes_train = pd.DataFrame()

for i in range(15):
    bbox_column = f'bbox_{i}'
    if bbox_column in df_train_class.columns:
        yolo_bboxes_train[bbox_column] = df_train_class.apply(\
            lambda row: convert_bbox_to_yolo(row[bbox_column], row['image_width'], row['image_height']) if not pd.isna(row[bbox_column]) else None,\
            axis=1)

yolo_bboxes_train.head()
```

```python
image_directory2 = '/content/leafdry/test/'
df_test_class['image_path'] = image_directory2 + df_test_class['img']
df_test_class['image_dimensions'] = df_test_class['image_path'].apply(get_image_dim)
df_test_class[['image_width', 'image_height']] = df_test_class['image_dimensions'].apply(pd.Series)

yolo_bboxes_test = pd.DataFrame()

for i in range(17):
    bbox_column = f'bbox_{i}'
    if bbox_column in df_test_class.columns:
        yolo_bboxes_test[bbox_column] = df_test_class.apply(\
            lambda row: convert_bbox_to_yolo(row[bbox_column], row['image_width'], row['image_height']) if not pd.isna(row[bbox_column]) else None,\
            axis=1)

yolo_bboxes_test.head()
```

```python
image_directory3 = '/content/leafdry/validation/'
df_validation_class['image_path'] = image_directory3 + df_validation_class['img']
df_validation_class['image_dimensions'] = df_validation_class['image_path'].apply(get_image_dim)
df_validation_class[['image_width', 'image_height']] = df_validation_class['image_dimensions'].apply(pd.Series)

yolo_bboxes_validation = pd.DataFrame()

for i in range(23):
    bbox_column = f'bbox_{i}'
    if bbox_column in df_validation_class.columns:
        yolo_bboxes_validation[bbox_column] = df_validation_class.apply(\
            lambda row: convert_bbox_to_yolo(row[bbox_column], row['image_width'], row['image_height']) if not pd.isna(row[bbox_column]) else None,\
            axis=1)

yolo_bboxes_validation.head()
```

```python
image_directory4 = '/content/leafdry/model_validation/model_validation_img/'
df_model_validation_class['image_path'] = image_directory4 + df_model_validation_class['img']
df_model_validation_class['image_dimensions'] = df_model_validation_class['image_path'].apply(get_image_dim)
df_model_validation_class[['image_width', 'image_height']] = df_model_validation_class['image_dimensions'].apply(pd.Series)

yolo_bboxes_model_validation = pd.DataFrame()

for i in range(12):
    bbox_column = f'bbox_{i}'
    if bbox_column in df_model_validation_class.columns:
        yolo_bboxes_model_validation[bbox_column] = df_model_validation_class.apply(\
            lambda row: convert_bbox_to_yolo(row[bbox_column], row['image_width'], row['image_height']) if not pd.isna(row[bbox_column]) else None,\
            axis=1)

yolo_bboxes_model_validation.head()
```

```python
annotation_train = yolo_bboxes_train.copy()
annotation_test = yolo_bboxes_test.copy()
annotation_validation = yolo_bboxes_validation.copy()
annotation_model_validation = yolo_bboxes_model_validation.copy()
```

```python
## 어노테이션 파일 생성 작업 1
  ## 대상 이미지 파일과 같은 이름으로 생성하기 위해 파일명 컬럼 복사
  ## 각각의 bbox 값에 할당된 class 정보 컬럼 복사
  ## yolo_bboxes 데이터프레임에 위의 두 컬럼 데이터를 합쳐, 어노테이션 생성용 최종 데이터프레임 저장

annotation_train['img'] = df_train_class['img']
annotation_test['img'] = df_test_class['img']
annotation_validation['img'] = df_validation_class['img']
annotation_model_validation['img'] = df_model_validation_class['img']

for i in range(15):
  class_column = f'class_{i}'
  if class_column in df_train_class.columns:
    annotation_train[class_column] = df_train_class[class_column]

for i in range(17):
  class_column = f'class_{i}'
  if class_column in df_test_class.columns:
    annotation_test[class_column] = df_test_class[class_column]

for i in range(23):
  class_column = f'class_{i}'
  if class_column in df_validation_class.columns:
    annotation_validation[class_column] = df_validation_class[class_column]

for i in range(12):
  class_column = f'class_{i}'
  if class_column in df_model_validation_class.columns:
    annotation_model_validation[class_column] = df_model_validation_class[class_column]
```

```python
annotation_train.to_csv('annotation_train.csv', encoding='utf-8', index=False)
annotation_test.to_csv('annotation_test.csv', encoding='utf-8', index=False)
annotation_validation.to_csv('annotation_validation.csv', encoding='utf-8', index=False)
annotation_model_validation.to_csv('annotation_model_validation.csv', encoding='utf-8', index=False)
```

```python
anno_train = pd.read_csv('/content/annotation_train.csv', encoding='utf-8')
anno_test = pd.read_csv('/content/annotation_test.csv', encoding='utf-8')
anno_validation = pd.read_csv('/content/annotation_validation.csv', encoding='utf-8')
anno_model_validation = pd.read_csv('/content/annotation_model_validation.csv', encoding='utf-8')

train_dir = '/content/leafdry/train/train_img/'
test_dir= '/content/leafdry/test/'
validation_dir = '/content/leafdry/validation/'
model_validation_dir = '/content/leafdry/model_validation/model_validation_img/'
```

```python
## 어노테이션 정보를 저장한 데이터프레임으로부터 각 이미지 데이터에 대한 어노테이션 파일 생성
  ## ['img']: 어노테이션 파일명 정보
  ## 어노테이션 txt 파일 구성: <class: 부호화한 클래스 라벨> <x: bbox[0]> <y: bbox[1]> <width: bbox[2]> <height:bbox[3]>

import ast
import os

classes = ['leaf-N', 'stem-N', 'flower-N', 'fruit-N', 'leaf-E', 'stem-E','flower-E', 'fruit-E', 'leaf-Y', 'stem-Y', 'flower-Y', 'fruit-Y']
class_to_index = {cls_name: index for index, cls_name in enumerate(classes)}

def create_annotation(df, annotations_dir):
  os.makedirs(annotations_dir, exist_ok=True)

  for index, row in df.iterrows():
    image_name = row['img'].rsplit('.', 1)[0]
    annotation_file = os.path.join(annotations_dir, f"{image_name}.txt")

    with open(annotation_file, 'w') as file:
      for i in range(25):
        bbox_col = f'bbox_{i}'
        class_col = f'class_{i}'

        if bbox_col in df.columns and class_col in df.columns and row[bbox_col] is not None:
          if isinstance(row[bbox_col], str):
            bbox = ast.literal_eval(row[bbox_col])
            class_name = row[class_col]

            class_index = class_to_index.get(class_name, -1)
            if class_index != -1:
                file.write(f"{class_index} {bbox[0]} {bbox[1]} {bbox[2]} {bbox[3]}\n")

annotation_prep = [anno_train, anno_test, anno_validation, anno_model_validation]
annotations_dir = [train_dir, test_dir, validation_dir, model_validation_dir]

for i in range(4):
    create_annotation(annotation_prep[i], annotations_dir[i])
```
